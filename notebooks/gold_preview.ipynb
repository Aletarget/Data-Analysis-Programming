# Gold Preview

This notebook previews the governance and storytelling Parquet outputs written to the Gold layer.

```python
from pathlib import Path
import pandas as pd

gold_dir = Path('../datalake_gold')
governance_files = sorted(gold_dir.glob('governance_*.parquet'))
storytelling_files = sorted(gold_dir.glob('storytelling_*.parquet'))
gov = pd.read_parquet(governance_files[-1]) if governance_files else pd.DataFrame()
story = pd.read_parquet(storytelling_files[-1]) if storytelling_files else pd.DataFrame()
gov.head(), story.head()
```

```python
if not gov.empty:
    display(gov)
else:
    print('No governance Parquet files found.')

if not story.empty:
    display(story)
else:
    print('No storytelling Parquet files found.')
```
