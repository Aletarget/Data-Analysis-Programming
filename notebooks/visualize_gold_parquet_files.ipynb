---

```python
from pathlib import Path
import pandas as pd

GOLD_PATH = Path("../datalake_gold")

datasets = [
    p for p in GOLD_PATH.rglob("*.parquet")
    if p.is_dir()
]

print(f"Datasets encontrados: {len(datasets)}")

for dataset in datasets:
    print("\n" + "=" * 120)
    print(dataset.relative_to(GOLD_PATH))

    df = pd.read_parquet(dataset)

    print(f"Rows: {len(df)}")
    print(f"Columns: {len(df.columns)}")

    display(df.head())
```
