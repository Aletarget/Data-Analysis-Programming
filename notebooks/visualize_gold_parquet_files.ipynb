## Visualice the .parquet files saved in datalake_gold 

Import libraries and declare paths

```python
import pandas as pd
from pathlib import Path

GOLD_BASE_PATH = "../datalake_gold"

# Ver todos los archivos disponibles
gold_files = sorted(Path(GOLD_BASE_PATH).glob("*.parquet"))
for f in gold_files:
    print(f.name)
```

---

```python
def ver_gold(nombre_parcial: str):
    gold_dir = Path(GOLD_BASE_PATH)
    files = sorted(gold_dir.glob(f"{nombre_parcial}*.parquet"), reverse=True)
    
    if not files:
        print(f"No se encontró archivo con prefijo '{nombre_parcial}'")
        return
    
    df = pd.read_parquet(files[0])
    print(f"\n{files[0].name}")
    print(f"Forma: {df.shape}")
    print(f"Columnas: {list(df.columns)}\n")
    display(df)
    return df

# Governance
gov_noticias = ver_gold("governance_noticias")
gov_tweets   = ver_gold("governance_tweets")

# Storytelling noticias
top_news        = ver_gold("storytelling_top_news")
comment_length  = ver_gold("storytelling_comment_length")
comments_flat   = ver_gold("storytelling_comments_flat")

# Storytelling tweets
top_tweets      = ver_gold("storytelling_top_tweets")
engagement_lang = ver_gold("storytelling_engagement_by_lang")
top_authors     = ver_gold("storytelling_top_authors")
tweets_flat     = ver_gold("storytelling_tweets_flat")
```
